#encoding labels

```python
columns= ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
le=LabelEncoder()
for i in columns:
  salary_train[i]=le.fit_transform(salary_train[i])
  salary_test[i]=le.fit_transform(salary_test[i])
```

```python
salary_train
```

```python
salary_test
```

#splitting data into train and test

```python
x_train=salary_train.iloc[:,0:13]
x_train
```

```python
x_test=salary_test.iloc[:,0:13]
x_test
```

```python
y_train=salary_train['Salary']
y_train
```

```python
y_test=salary_test['Salary']
y_test
```

#Standardising data

```python
std=StandardScaler()
x_train=pd.DataFrame(std.fit_transform(x_train))
x_train
```

```python
x_test=pd.DataFrame(std.fit_transform(x_test))
x_test
```

```python
y_train=le.fit_transform(y_train)
y_train
```

```python
y_test=le.fit_transform(y_test)
y_test
```

#Naive bayes model 

```python
ga=GaussianNB()
```

```python
ga.fit(x_train,y_train)
```

```python
y_pred_train=ga.predict(x_train)
y_pred_train

```

```python
y_pred_test=ga.predict(x_test)
y_pred_test
```

```python
accuracy_train=accuracy_score(y_train,y_pred_train)*100
accuracy_train
```

```python
accuracy_test=accuracy_score(y_test,y_pred_test)*100
accuracy_test
```
